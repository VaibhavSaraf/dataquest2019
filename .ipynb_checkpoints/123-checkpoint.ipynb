```python
def extracting_num(df_raw):
  df = df_raw.copy()
  df['opened_by'] = df['opened_by'].str.split(expand=True)[2]
  df['location'] = df['location'].str.split(expand=True)[1]
  df['category'] = df['category'].str.split(expand=True)[1]
  df['subcategory'] = df['subcategory'].str.split(expand=True)[1]
  df['impact'] = df['impact'].str.split(expand=True)[0]
  df['urgency'] = df['urgency'].str.split(expand=True)[0]
  df['priority'] = df['priority'].str.split(expand=True)[0]
  df['assigned_to'] = df['assigned_to'].str.split(expand=True)[1]
  return df
```

```python
train_raw = extracting_num(train_raw)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
df['knowledge']= label_encoder.fit_transform(df['knowledge'])
df['opened_atIs_month_end']= label_encoder.fit_transform(df['opened_atIs_month_end'])
df['opened_atIs_month_start']= label_encoder.fit_transform(df['opened_atIs_month_start'])
df['opened_atIs_quarter_end']= label_encoder.fit_transform(df['opened_atIs_quarter_end'])
df['opened_atIs_quarter_start']= label_encoder.fit_transform(df['opened_atIs_quarter_start'])
df['contact_type']= label_encoder.fit_transform(df['contact_type'])
```

```python
df.isnull().sum()
```

```python
X = df.copy()
```

```python
X.drop(columns='target_days',inplace=True)
```

```python
y = df['target_days']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
def print_score(m):
    print(metrics.mean_absolute_error(m.predict(X_train),y_train),metrics.mean_absolute_error(m.predict(X_valid),y_valid))
```

```python
from sklearn.neighbors import KNeighborsClassifier
n = KNeighborsClassifier()
n.fit(X_train,y_train)
print_score(n)
```

```python
from sklearn.ensemble import RandomForestRegressor , RandomForestClassifier
m = RandomForestClassifier(n_jobs=-1,random_state=3,min_samples_leaf=1)
m.fit(X_train,y_train)
print_score(m)
```

```python
k = RandomForestRegressor(n_jobs=-1,random_state=3,min_samples_leaf=1)
k.fit(X_train,y_train)
print_score(k)
```
